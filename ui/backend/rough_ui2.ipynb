```python
# start_date = "2000-01-01"
# end_date = "2024-12-01"
DESIGNATION_PATH = '../../data_ui2/interim/Designation.csv'


def generate_odi_data_training(start_date,end_date):

    json_dir = "../../data_ui2/raw/cricksheet_data"  
    output_csv = f"../../data_ui2/interim/ODI_MatchWise_{end_date}.csv" 
    ODI_MatchWise(json_dir,output_csv,start_date,end_date)

    input_file = f"../../data_ui2/interim/ODI_MatchWise_{end_date}.csv"
    output_file = f"../../data_ui2/interim/ODI_MatchWise_fantasy_points_{end_date}.csv"
    add_fantasy_points_odi(input_file, output_file)

    input_file = f'../../data_ui2/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/ODI/career_odi_{end_date}.csv'
    career_ODI(input_file, output_file)

    input_file = f'../../data_ui2/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/ODI/recent_odi_{end_date}.csv'
    recent_ODI(input_file, output_file)

    input_file = f'../../data_ui2/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/ODI/venue_odi_{end_date}.csv'
    venue_ODI(input_file, output_file)

    input_file = f'../../data_ui2/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/ODI/player_match_data_odi_{end_date}.csv'
    process_player_match_data_odi(DESIGNATION_PATH,input_file, output_file)



def generate_t20_data_training(start_date,end_date):

    json_dir = "../../data_ui2/raw/cricksheet_data"  
    output_csv = f"../../data_ui2/interim/T20_MatchWise_{end_date}.csv" 
    T20_MatchWise(json_dir,output_csv,start_date,end_date)

    input_file = f'../../data_ui2/interim/T20_MatchWise_{end_date}.csv'
    output_file = f'../../data_ui2/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    add_fantasy_points_t20(input_file, output_file)

    input_file = f'../../data_ui2/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/T20/career_t20_{end_date}.csv'
    career_T20(input_file, output_file)

    input_file = f'../../data_ui2/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/T20/recent_t20_{end_date}.csv'
    recent_T20(input_file, output_file)

    input_file = f'../../data_ui2/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/T20/venue_t20_{end_date}.csv'
    venue_T20(input_file, output_file)

    input_file = f'../../data_ui2/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/T20/player_match_data_t20_{end_date}.csv'
    process_player_match_data_t20(DESIGNATION_PATH,input_file, output_file)



def generate_test_data_training(start_date,end_date):

    json_dir = "../../data_ui2/raw/cricksheet_data"  
    output_csv = f"../../data_ui2/interim/Test_MatchWise_{end_date}.csv" 
    Test_MatchWise(json_dir,output_csv,start_date,end_date)

    input_file = f'../../data_ui2/interim/Test_MatchWise_{end_date}.csv'
    output_file = f'../../data_ui2/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    add_fantasy_points_test(input_file, output_file)

    input_file = f'../../data_ui2/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/Test/career_test_{end_date}.csv'
    career_Test(input_file, output_file)

    input_file = f'../../data_ui2/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/Test/recent_test_{end_date}.csv'
    recent_Test(input_file, output_file)

    input_file = f'../../data_ui2/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/Test/venue_test_{end_date}.csv'
    venue_Test(input_file, output_file)

    input_file = f'../../data_ui2/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/Test/player_match_data_test_{end_date}.csv'
    process_player_match_data_test(DESIGNATION_PATH,input_file, output_file)


def generate_odi_data_testing(start_date,end_date):

    json_dir = "../../data_ui2/raw/cricksheet_data"  
    output_csv = f"../../data_ui2/test_data/interim/ODI_MatchWise_{end_date}.csv" 
    ODI_MatchWise(json_dir,output_csv,start_date,end_date)

    input_file = f"../../data_ui2/test_data/interim/ODI_MatchWise_{end_date}.csv"
    output_file = f"../../data_ui2/test_data/interim/ODI_MatchWise_fantasy_points_{end_date}.csv"
    add_fantasy_points_odi(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/ODI/career_odi_{end_date}.csv'
    career_ODI(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/ODI/recent_odi_{end_date}.csv'
    recent_ODI(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/ODI/venue_odi_{end_date}.csv'
    venue_ODI(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/ODI_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/ODI/player_match_data_odi_{end_date}.csv'
    process_player_match_data_odi(DESIGNATION_PATH,input_file, output_file)



def generate_t20_data_testing(start_date,end_date):

    json_dir = "../../data_ui2/raw/cricksheet_data"  
    output_csv = f"../../data_ui2/test_data/interim/T20_MatchWise_{end_date}.csv" 
    T20_MatchWise(json_dir,output_csv,start_date,end_date)

    input_file = f'../../data_ui2/test_data/interim/T20_MatchWise_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    add_fantasy_points_t20(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/T20/career_t20_{end_date}.csv'
    career_T20(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/processed/T20/recent_t20_{end_date}.csv'
    recent_T20(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/T20/venue_t20_{end_date}.csv'
    venue_T20(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/T20_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/T20/player_match_data_t20_{end_date}.csv'
    process_player_match_data_t20(DESIGNATION_PATH,input_file, output_file)



def generate_test_data_testing(start_date,end_date):

    json_dir = "../../data_ui2/raw/cricksheet_data"  
    output_csv = f"../../data_ui2/test_data/interim/Test_MatchWise_{end_date}.csv" 
    Test_MatchWise(json_dir,output_csv,start_date,end_date)

    input_file = f'../../data_ui2/test_data/interim/Test_MatchWise_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    add_fantasy_points_test(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/Test/career_test_{end_date}.csv'
    career_Test(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/Test/recent_test_{end_date}.csv'
    recent_Test(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/Test/venue_test_{end_date}.csv'
    venue_Test(input_file, output_file)

    input_file = f'../../data_ui2/test_data/interim/Test_MatchWise_fantasy_points_{end_date}.csv'
    output_file = f'../../data_ui2/test_data/processed/Test/player_match_data_test_{end_date}.csv'
    process_player_match_data_test(DESIGNATION_PATH,input_file, output_file)


def generate_training_data_for_retraining(start_date,end_date):

    generate_odi_data_training(start_date,end_date)
    generate_t20_data_training(start_date,end_date)
    generate_test_data_training(start_date,end_date)
    
def generate_testing_data_for_retraining(start_date,end_date):

    generate_odi_data_testing(start_date,end_date)
    generate_t20_data_testing(start_date,end_date)
    generate_test_data_testing(start_date,end_date)



def train_model_odi(start_date, end_date):
    # Paths
    file1_path = f'../../data_ui2/processed/ODI/career_odi_{end_date}.csv'
    file2_path = f'../../data_ui2/processed/ODI/recent_odi_{end_date}.csv'
    file3_path = f'../../data_ui2/processed/ODI/venue_odi_{end_date}.csv'
    file4_path = f'../../data_ui2/processed/ODI/player_match_data_odi_{end_date}.csv'

    # Load data
    df1 = pd.read_csv(file1_path, dtype=career_dict)
    df2 = pd.read_csv(file2_path, dtype=recent_dict)
    df3 = pd.read_csv(file3_path, dtype=venue_dict)
    df4 = pd.read_csv(file4_path, dtype=player_dict, parse_dates=["date"])

    # Merge datasets
    merged_df = pd.merge(df1, df2, on=['match_id', 'player_id'], how='inner')
    merged_df = pd.merge(merged_df, df3, on=['match_id', 'player_id'], how='inner')
    merged_df = pd.merge(merged_df, df4, on=['match_id', 'player_id'], how='inner')

    # Process data
    merged_df['date'] = pd.to_datetime(merged_df['date'])
    columns_to_drop = ['match_id', 'player_id', 'player_name', 'fantasy_points', 'team_name', 'role', 'date', 'venue']
    X_train = merged_df.drop(columns=columns_to_drop)
    y_train = merged_df['fantasy_points']

    # Scale the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)

    # Train the model
    model = HuberRegressor(max_iter=500)  # Increased max_iter
    model.fit(X_train_scaled, y_train)

    # Save the model
    filename = f"../../model_artifacts_ui2/ODI/model_ODI_{end_date}.pkl"
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    with open(filename, 'wb') as f:
        pickle.dump(model, f)

```
