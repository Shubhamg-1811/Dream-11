```python
import pandas as pd

# Load the previously saved CSV containing player-wise errors
error_df = pd.read_csv('player_error_details.csv')

# Aggregate the data by 'match_id'
matchwise_summary = error_df.groupby('match_id').agg(
    actual_sum_fantasy_points=('actual_fantasy_points', 'sum'),
    predicted_sum_fantasy_points=('predicted_fantasy_points', 'sum')
).reset_index()

# Calculate error and percentage_error
matchwise_summary['error'] = matchwise_summary['predicted_sum_fantasy_points'] - matchwise_summary['actual_sum_fantasy_points']
matchwise_summary['percentage_error'] = matchwise_summary.apply(
    lambda row: abs(row['error']) / abs(row['actual_sum_fantasy_points']) * 100 if row['actual_sum_fantasy_points'] != 0 else -1,
    axis=1
)

# Save the matchwise summary to a new CSV
matchwise_summary.to_csv('matchwise_error_summary.csv', index=False)

# Print the matchwise error summary
print("Matchwise Error Summary:")
print(matchwise_summary)
```

```python
import pandas as pd

# Input File
predicted_odi_teams_file = "/Users/chaitanyatandon/Desktop/DREAM_11_v2/Dream-11/Code/Model/predicted_odi_teams.csv"  # Path to the predicted_odi_teams.csv

# Output File
output_file = "predicted_dream_team.csv"  # Path to save the output CSV

# Read Data
odi_teams_data = pd.read_csv(predicted_odi_teams_file)

# Columns for Output
output_columns = [
    "match_id",
    "player1_id", "player1_predicted_fantasy_points",
    "player2_id", "player2_predicted_fantasy_points",
    "player3_id", "player3_predicted_fantasy_points",
    "player4_id", "player4_predicted_fantasy_points",
    "player5_id", "player5_predicted_fantasy_points",
    "player6_id", "player6_predicted_fantasy_points",
    "player7_id", "player7_predicted_fantasy_points",
    "player8_id", "player8_predicted_fantasy_points",
    "player9_id", "player9_predicted_fantasy_points",
    "player10_id", "player10_predicted_fantasy_points",
    "player11_id", "player11_predicted_fantasy_points"
]

# Prepare Output Data
output_data = []

# Function to calculate the predicted dream team
def calculate_predicted_dream_team(players, match_group):
    try:
        selected_players = []

        # Step 1: Pick one player from each role based on predicted fantasy points
        for role in ["Batsman", "Bowler", "Wicket-Keeper", "All-Rounder"]:
            role_players = [p for p in players if p["role"] == role]
            if role_players:
                selected_player = max(role_players, key=lambda x: x["predicted_fantasy_points"])
                selected_players.append(selected_player)

        # Step 2: Select remaining players to make a team of 11
        remaining_players = [p for p in players if p not in selected_players]
        remaining_players.sort(key=lambda x: x["predicted_fantasy_points"], reverse=True)

        while len(selected_players) < 11:
            next_player = remaining_players.pop(0)
            selected_players.append(next_player)

        # Step 3: Ensure team diversity (No more than 7 players from the same team)
        team_counts = pd.DataFrame(selected_players)["team_name"].value_counts()
        if team_counts.max() == 11:
            # Replace the lowest fantasy point player with the highest fantasy point player from the other team
            other_team = [team for team in match_group["team_name"].unique() if team not in team_counts.index][0]
            lowest_fantasy_player = min(selected_players, key=lambda x: x["predicted_fantasy_points"])
            replacement_player = max(
                [p for p in players if p["team_name"] == other_team and p not in selected_players],
                key=lambda x: x["predicted_fantasy_points"]
            )
            selected_players.remove(lowest_fantasy_player)
            selected_players.append(replacement_player)

        # Step 4: Return selected players for the match
        return selected_players

    except Exception as e:
        print(f"Error in calculating dream team: {e}")
        return None

# Process the data for each match and generate the output
for match_id, match_group in odi_teams_data.groupby("match_id"):
    # Get the list of players in the match group as dictionaries
    players = match_group.to_dict(orient="records")
    
    # Calculate the predicted dream team for the match
    selected_players = calculate_predicted_dream_team(players, match_group)
    
    # If a valid team is formed (i.e., 11 players selected), extract the relevant data
    if selected_players and len(selected_players) == 11:
        output_row = [match_id]
        
        # Add player details to the output row
        for i in range(11):
            player = selected_players[i]
            output_row.append(player["player_id"])
            output_row.append(player["predicted_fantasy_points"])
        
        # Append the row to the output data
        output_data.append(output_row)

# Create a DataFrame from the output data
output_df = pd.DataFrame(output_data, columns=output_columns)

# Save the output to a CSV file
output_df.to_csv(output_file, index=False)

print(f"Predicted dream teams saved to {output_file}")
```
