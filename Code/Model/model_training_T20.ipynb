```python
import pandas as pd

# Paths to the CSV files
file_path1 = '/Users/chaitanyatandon/Desktop/DREAM_11_v2/Dream-11/CSVs/Sohit/t20_match_wise_cumulated.csv'
file_path2 = '/Users/chaitanyatandon/Desktop/DREAM_11_v2/Dream-11/CSVs/Sohit/recent_form_t20.csv'

# Load the CSV files into pandas DataFrames
df_t20_match_wise_cumulated = pd.read_csv(file_path1)
df_recent_form_t20 = pd.read_csv(file_path2)

df1=df_recent_form_t20
df2=df_t20_match_wise_cumulated

# Display the DataFrames to ensure they are loaded correctly
print("T20 Match Wise Cumulated DataFrame:")
print(df_t20_match_wise_cumulated.head())

print("\nRecent Form T20 DataFrame:")
print(df_recent_form_t20.head())
```

```python
# Assuming your DataFrames are named df1 and df2
# Replace 'Match_ID' and 'Player_ID' with the actual column names if different
merged_df = pd.merge(df1, df2, on=['match_id', 'player_id'], how='inner')

# Display the merged DataFrame
print(merged_df.head())
```

```python
# Assuming the DataFrame is named df
fields = merged_df.columns.tolist()

# Display the fields
print("List of fields in the DataFrame:")
print(fields)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import HuberRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

df= merged_df
# Define the input features and target variable
input_features = [
    'batsman_100s_t20', 'batsman_50s_t20', 'batsman_30s_t20',
    'batsman_average_runs_per_match_t20', 'batsman_strike_rate_t20',
    'batsman_total_runs_t20', 'bowler_average_t20', 'bowler_economy_rate_t20',
    'bowler_wickets_t20', 'fielder_total_catches_t20', 'fielder_total_runouts_t20',
    'recent_total_runs_t20', 'recent_average_t20', 'recent_strike_rate_t20',
    'recent_4s_t20', 'recent_6s_t20', 'recent_30s_t20', 'recent_50s_t20',
    'recent_100s_t20', 'recent_economy_rate_t20', 'recent_bowler_average_t20',
    'recent_total_wickets_t20', 'recent_total_catches_t20', 'recent_runouts_t20'
]

target_variable = 'fantasy_points_x'

# Ensure the DataFrame contains required columns
identifier_columns = ['match_id', 'player_id']
required_columns = input_features + [target_variable] + identifier_columns
missing_columns = [col for col in required_columns if col not in df.columns]
if missing_columns:
    raise ValueError(f"The DataFrame is missing the following columns: {missing_columns}")


# Drop rows with missing values in the required columns
df = df.dropna(subset=required_columns)

# Define input (X) and target (y)
X = df[input_features]
y = df[target_variable]

# Retain identifiers for output
X_identifiers = df[identifier_columns]

# Sequential split (80% train, 20% test)
split_index = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
X_test_identifiers = X_identifiers.iloc[split_index:].reset_index(drop=True)

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train Huber Regressor
print("Training Huber Regressor...")
huber = HuberRegressor()
huber.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = huber.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Absolute Percentage Error (MAPE)
non_zero_indices = y_test != 0  # Avoid division by zero
y_test_non_zero = y_test[non_zero_indices]
y_pred_non_zero = y_pred[non_zero_indices]
percentage_errors = np.abs((y_test_non_zero - y_pred_non_zero) / y_test_non_zero) * 100
mape = np.mean(percentage_errors)

# Print evaluation metrics
print("\nHuber Regression Results:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Save player-wise predictions to a CSV
playerwise_output = pd.DataFrame({
    'match_id': X_test_identifiers['match_id'],
    'player_id': X_test_identifiers['player_id'],
    'Actual_Fantasy_Points': y_test.values,
    'Predicted_Fantasy_Points': y_pred,
    'Absolute_Error': np.abs(y_pred - y_test.values),
    'Percentage_Error': np.abs((y_pred - y_test.values) / y_test.values) * 100  # Percentage error
}).reset_index(drop=True)

playerwise_output_file = 'playerwise_predictions.csv'
playerwise_output.to_csv(playerwise_output_file, index=False)
print(f"Player-wise predictions saved to '{playerwise_output_file}'")

# Group by match_id and calculate summary statistics
grouped_output = playerwise_output.groupby('match_id').agg({
    'Actual_Fantasy_Points': ['mean', 'sum', 'min', 'max'],  # Summary for actual points
    'Predicted_Fantasy_Points': ['mean', 'sum', 'min', 'max'],  # Summary for predicted points
    'Absolute_Error': ['mean', 'sum'],  # Summary for absolute error
})

# Calculate Percentage Error in Total Points
grouped_output['Total_Points_Percentage_Error'] = (
    (grouped_output[('Predicted_Fantasy_Points', 'sum')] - grouped_output[('Actual_Fantasy_Points', 'sum')]).abs()
    / grouped_output[('Actual_Fantasy_Points', 'sum')]
) * 100

# Flatten multi-level columns for better readability
grouped_output.columns = ['_'.join(col) if isinstance(col, tuple) else col for col in grouped_output.columns]
grouped_output = grouped_output.reset_index()

# Save grouped predictions
grouped_output_file = 'grouped_predictions.csv'
grouped_output.to_csv(grouped_output_file, index=False)
print(f"Grouped predictions saved to '{grouped_output_file}'")
```
