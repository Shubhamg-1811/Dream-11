```python
import pandas as pd
df = pd.read_csv('../../CSVs/match_data2.csv', low_memory=False)
df
```

```python
# Check the data types of all values in the player_id column
non_string_ids = df[df['player_id'].apply(lambda x: not isinstance(x, str))]

# Print the rows where player_id is not a string
# print(non_string_ids)

# Count how many rows have non-string player_id
print(f"Number of non-string player_ids: {len(non_string_ids)}")
```

```python
# Filter rows where player_name is "JR Hopes"
jr_hopes_rows = df[df['player_name'] == 'JR Hopes']

# Display the filtered rows
print(jr_hopes_rows)

# Count the number of rows with player_name as "JR Hopes"
print(f"Number of rows where player_name is JR Hopes: {len(jr_hopes_rows)}")
```
